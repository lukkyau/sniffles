```python
import pandas as pd
import sqlite3
import random, math
from numpy.random import permutation

conn = sqlite3.connect("signal_strengths.db")
df = pd.read_sql_query("select * from signals;", conn)

columns = list(df.filter(regex='MAC'))
columns.append('collected')
print(columns)

df=df.drop(columns,axis=1)
df.head()

# Create train/test dataset
random_indices = permutation(df.index)
test_cutoff = math.floor(len(df)/3)
test = df.loc[random_indices[1:test_cutoff]]
train = df.loc[random_indices[test_cutoff:]]

test.head()
train.head()
```
