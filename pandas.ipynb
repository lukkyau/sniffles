```python
import pandas as pd
import sqlite3

conn = sqlite3.connect("signal_strengths.db")
df = pd.read_sql_query("select * from signals;", conn,index_col='collected')
df.index
df.dtypes
df.head() # Show a few examples
```

```python
# Based on the strengths of two access points, can separate out bedroom and living room
cmap = {'room': 'red', 'living': 'blue'}
df.plot(kind='scatter',x='AP2',y='AP5', color=[cmap.get(c, 'black') for c in df.location])
plt.show()
```

```python
import random, math
from numpy.random import permutation

# Create test and train/learn datasets
random_indices = permutation(df.index)
test_cutoff = math.floor(len(df)/3)
test = df.loc[random_indices[1:test_cutoff]]
train = df.loc[random_indices[test_cutoff:]]

# Use scikit-learn to predict
from sklearn.neighbors import KNeighborsClassifier
x_columns=['AP1','AP2','AP3','AP4']
y_column=['location']
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(train[x_columns],train[y_column].values.ravel())
predictions = knn.predict(test[x_columns])

# Check error rate
actual = test[y_column]
actual['predicted']=predictions
actual
```
